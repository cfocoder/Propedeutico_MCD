```python
# Importar las librerías necesarias

from IPython.display import display, Math
import sympy  ##  Es una librería de Python para matemáticas simbólicas, lo que permite trabajar 
# con expresiones matemáticas de manera simbólica, similar a cómo se haría en papel.
# Es ideal para problemas de álgebra, cálculo y ecuaciones diferenciales, entre otros.
sympy.init_printing(use_unicode=True) # Configuración para que Simpy muestre las expresiones matemáticas de forma más legible 
```

```python
# PRIMER EJERCICIO

print("Calculando la derivada de f(x) = (6x^2 + 7x)^2 usando la regla de la cadena:\n")

# 1. Definir el símbolo y la función original
x = sympy.symbols('x')

# Paso 1: Identificar la función interna u(x)
u_expr = 6*x**2 + 7*x
print(f"Paso 1: Sea la función interna u(x) = {u_expr}")

# Paso 2: Definir la función externa h(u) en términos de un nuevo símbolo 'u'
# y cómo f(x) se relaciona con h(u): f(x) = h(u(x))
u_sym = sympy.symbols('u') # Símbolo para la función interna
h_expr_u = u_sym**2
print(f"Paso 2: La función externa es h(u) = {h_expr_u}")
print(f"        Así, f(x) = h(u(x)) donde u(x) = {u_expr}")

# Paso 3: Calcular la derivada de la función interna con respecto a x (du/dx)
du_dx = sympy.diff(u_expr, x)
print(f"Paso 3: Derivada de la función interna: du/dx = {du_dx}")

# Paso 4: Calcular la derivada de la función externa con respecto a u (dh/du)
dh_du = sympy.diff(h_expr_u, u_sym)
print(f"Paso 4: Derivada de la función externa respecto a u: dh/du = {dh_du}")

# Paso 5: Aplicar la regla de la cadena: f'(x) = dh/du(u(x)) * du/dx
# Sustituir u_expr (la expresión de u en términos de x) de nuevo en dh_du
dh_du_evaluated_at_u_of_x = dh_du.subs(u_sym, u_expr)
print(f"Paso 5a: Sustituir u(x) en dh/du para obtener dh/du(u(x)) = {dh_du_evaluated_at_u_of_x}")

# Multiplicar dh/du(u(x)) por du/dx
f_prime = dh_du_evaluated_at_u_of_x * du_dx
print(f"Paso 5b: Aplicar la regla de la cadena f'(x) = dh/du(u(x)) * du/dx")
print(f"         f'(x) = ({dh_du_evaluated_at_u_of_x}) * ({du_dx})")

# Resultado final de la aplicación de la regla de la cadena (antes de cualquier expansión)
print(f"\nResultado final aplicando la regla de la cadena:")
print(f"f'(x) = {f_prime}")

# Mostrar la forma expandida 
f_prime_expanded = sympy.expand(f_prime)
print(f"\nForma expandida (simplificada) del resultado:")
print(f"f'(x) = {f_prime_expanded}")
```

```python
# SEGUNDO EJERCICIO

# Calculando la derivada de g(x) = e^(1 - cos(x)) usando la regla de la cadena.

# 0. Definir el símbolo y la función 
x = sympy.symbols('x')

print(f"Calculando la primera derivada de g(x) = e^(1 - cos(x)) usando la regla de la cadena:\n")

# Paso 1: Identificar la función interna u(x)
u_expr = 1 - sympy.cos(x)
print(f"Paso 1: Sea la función interna u(x) = {u_expr}")

# Paso 2: Definir la función externa h(u) en términos de un nuevo símbolo 'u'
# y cómo g(x) se relaciona con h(u): g(x) = h(u(x))
u_sym = sympy.symbols('u') # Símbolo para la función interna
h_expr_u = sympy.exp(u_sym) # e^u
print(f"Paso 2: La función externa es h(u) = {h_expr_u}")
print(f"        Así, g(x) = h(u(x)) donde u(x) = {u_expr}")

# Paso 3: Calcular la derivada de la función interna con respecto a x (du/dx)
du_dx = sympy.diff(u_expr, x)
print(f"Paso 3: Derivada de la función interna: du/dx = {du_dx}")

# Paso 4: Calcular la derivada de la función externa con respecto a u (dh/du)
dh_du = sympy.diff(h_expr_u, u_sym)
print(f"Paso 4: Derivada de la función externa respecto a u: dh/du = {dh_du}")

# Paso 5: Aplicar la regla de la cadena: g'(x) = dh/du(u(x)) * du/dx
# Paso 5a: Sustituir u_expr (la expresión de u en términos de x) de nuevo en dh_du
dh_du_evaluated_at_u_of_x = dh_du.subs(u_sym, u_expr)
print(f"Paso 5a: Sustituir u(x) en dh/du para obtener dh/du(u(x)) = {dh_du_evaluated_at_u_of_x}")

# Paso 5b: Multiplicar dh/du(u(x)) por du/dx
g_prime = dh_du_evaluated_at_u_of_x * du_dx
print(f"Paso 5b: Aplicar la regla de la cadena g'(x) = dh/du(u(x)) * du/dx")
print(f"         g'(x) = ({dh_du_evaluated_at_u_of_x}) * ({du_dx})")

# Resultado final de la aplicación de la regla de la cadena (antes de cualquier expansión)
print(f"\nResultado final aplicando la regla de la cadena:")
print(f"g'(x) = {g_prime}")

# Forma expandida (simplificada) del resultado:
g_prime_expanded = sympy.expand(g_prime) # o sympy.simplify(g_prime)
print(f"\nForma expandida (simplificada) del resultado:")
print(f"g'(x) = {g_prime_expanded}")
```

```python
# TERCER EJERCICIO

print(f"Calculando la primera derivada de g(z) = sin(z^6) + sin^6(z):\n")
print(f"La función es una suma de dos términos: A(z) = sin(z^6) y B(z) = sin^6(z).")
print(f"Por lo que calcularemos A'(z) y B'(z) por separado usando la regla de la cadena, y luego g'(z) = A'(z) + B'(z).\n")

# 0. Definir el símbolo
z = sympy.symbols('z')

# --- Parte 1: Derivada de A(z) = sin(z^6) ---
print("--- Parte 1: Calculando la derivada de A(z) = sin(z^6) ---")

# Paso 1.1: Identificar la función interna u_A(z) para A(z)
u_A_expr = z**6
print(f"Paso 1.1: Para A(z), sea la función interna u_A(z) = {u_A_expr}")

# Paso 1.2: Definir la función externa h_A(u_A) para A(z)
u_A_sym = sympy.symbols('u_A') # Símbolo para la función interna de A
h_A_expr_u = sympy.sin(u_A_sym)
print(f"Paso 1.2: La función externa para A(z) es h_A(u_A) = {h_A_expr_u}")
print(f"          Así, A(z) = h_A(u_A(z)) donde u_A(z) = {u_A_expr}")

# Paso 1.3: Calcular la derivada de la función interna u_A(z) (du_A/dz)
du_A_dz = sympy.diff(u_A_expr, z)
print(f"Paso 1.3: Derivada de la función interna u_A(z): du_A/dz = {du_A_dz}")

# Paso 1.4: Calcular la derivada de la función externa h_A(u_A) (dh_A/du_A)
dh_A_du = sympy.diff(h_A_expr_u, u_A_sym)
print(f"Paso 1.4: Derivada de la función externa h_A(u_A) respecto a u_A: dh_A/du_A = {dh_A_du}")

# Paso 1.5: Aplicar la regla de la cadena para A'(z) = dh_A/du_A(u_A(z)) * du_A/dz
# Paso 1.5a: Sustituir u_A(z) en dh_A/du_A
dh_A_du_evaluated_at_u_A_of_z = dh_A_du.subs(u_A_sym, u_A_expr)
print(f"Paso 1.5a: Sustituir u_A(z) en dh_A/du_A para obtener dh_A/du_A(u_A(z)) = {dh_A_du_evaluated_at_u_A_of_z}")

# Paso 1.5b: Multiplicar
A_prime = dh_A_du_evaluated_at_u_A_of_z * du_A_dz
print(f"Paso 1.5b: Aplicar la regla de la cadena A'(z) = dh_A/du_A(u_A(z)) * du_A/dz")
print(f"           A'(z) = ({dh_A_du_evaluated_at_u_A_of_z}) * ({du_A_dz})")
print(f"           A'(z) = {A_prime}\n")


# --- Parte 2: Derivada de B(z) = sin^6(z) ---
# Recordar que sin^6(z) es (sin(z))^6
print("--- Parte 2: Calculando la derivada de B(z) = sin^6(z) ---")

# Paso 2.1: Identificar la función interna v_B(z) para B(z)
v_B_expr = sympy.sin(z)
print(f"Paso 2.1: Para B(z), sea la función interna v_B(z) = {v_B_expr}")

# Paso 2.2: Definir la función externa k_B(v_B) para B(z)
v_B_sym = sympy.symbols('v_B') # Símbolo para la función interna de B
k_B_expr_v = v_B_sym**6
print(f"Paso 2.2: La función externa para B(z) es k_B(v_B) = {k_B_expr_v}")
print(f"          Así, B(z) = k_B(v_B(z)) donde v_B(z) = {v_B_expr}")

# Paso 2.3: Calcular la derivada de la función interna v_B(z) (dv_B/dz)
dv_B_dz = sympy.diff(v_B_expr, z)
print(f"Paso 2.3: Derivada de la función interna v_B(z): dv_B/dz = {dv_B_dz}")

# Paso 2.4: Calcular la derivada de la función externa k_B(v_B) (dk_B/dv_B)
dk_B_dv = sympy.diff(k_B_expr_v, v_B_sym)
print(f"Paso 2.4: Derivada de la función externa k_B(v_B) respecto a v_B: dk_B/dv_B = {dk_B_dv}")

# Paso 2.5: Aplicar la regla de la cadena para B'(z) = dk_B/dv_B(v_B(z)) * dv_B/dz
# Paso 2.5a: Sustituir v_B(z) en dk_B/dv_B
dk_B_dv_evaluated_at_v_B_of_z = dk_B_dv.subs(v_B_sym, v_B_expr)
print(f"Paso 2.5a: Sustituir v_B(z) en dk_B/dv_B para obtener dk_B/dv_B(v_B(z)) = {dk_B_dv_evaluated_at_v_B_of_z}")

# Paso 2.5b: Multiplicar
B_prime = dk_B_dv_evaluated_at_v_B_of_z * dv_B_dz
print(f"Paso 2.5b: Aplicar la regla de la cadena B'(z) = dk_B/dv_B(v_B(z)) * dv_B/dz")
print(f"           B'(z) = ({dk_B_dv_evaluated_at_v_B_of_z}) * ({dv_B_dz})")
print(f"           B'(z) = {B_prime}\n")

# --- Parte 3: Sumar las derivadas A'(z) y B'(z) ---
print("--- Parte 3: Sumando las derivadas para obtener g'(z) ---")
g_prime = A_prime + B_prime
print(f"Paso 3.1: g'(z) = A'(z) + B'(z)")
print(f"           g'(z) = ({A_prime}) + ({B_prime})")

# Resultado final de la aplicación de la regla de la cadena
print(f"\nResultado final aplicando la regla de la cadena:")
print(f"g'(z) = {g_prime}")

# Forma expandida (simplificada) del resultado:
g_prime_expanded = sympy.expand(g_prime)
print(f"\nForma expandida (simplificada) del resultado:")
print(f"g'(z) = {g_prime_expanded}")
```

```python
# PRIMER EJERCICIO   

# --- PASO 1: Definir el símbolo y la función ---
print("--- PASO 1: Definir el símbolo y la función ---")
z = sympy.symbols('z')
h_expr = 4*z**3 - 3*z**2 + 9*z + 12
display(Math(f"h(z) = {sympy.latex(h_expr)}"))

# --- PASO 2: Determinar el dominio de la función ---
# h(z) es un polinomio, por lo que está definido para todos los números reales.
print("\n--- PASO 2: Determinar el dominio de la función ---")
domain_h = sympy.S.Reals # El conjunto de todos los números reales
display(Math(f"\\text{{Dominio de }} h(z): D = {sympy.latex(domain_h)}"))

# --- PASO 3: Calcular la primera derivada (h'(z)) ---
print("\n--- PASO 3: Calcular la primera derivada h'(z) ---")
h_prime = sympy.diff(h_expr, z)
display(Math(f"h'(z) = {sympy.latex(h_prime)}"))

# --- PASO 4: Encontrar los puntos críticos ---
# Los puntos críticos ocurren donde h'(z) = 0 o donde h'(z) no existe.
print("\n--- PASO 4: Encontrar los puntos críticos ---")

# a) Puntos donde h'(z) = 0
print("\na) Puntos donde h'(z) = 0:")
# Usamos sympy.solve para encontrar las raíces de h_prime
# Especificamos domain=sympy.S.Reals para buscar solo soluciones reales.
critical_points_h_prime_zero_set = sympy.solveset(h_prime, z, domain=sympy.S.Reals)

# Convertir el conjunto de soluciones (que será EmptySet si no hay reales) a una lista
critical_points_h_prime_zero = list(critical_points_h_prime_zero_set)

if not critical_points_h_prime_zero:
    print("No hay soluciones reales para h'(z) = 0.")
try:
    display(Math(f"\\text{{Soluciones reales para }} h'(z) = 0: z = {sympy.latex(critical_points_h_prime_zero)}"))
except NameError:
    print(f"Soluciones reales para h'(z) = 0: z = {critical_points_h_prime_zero}")

# b) Puntos donde h'(z) no está definida
# h'(z) = 12*z**2 - 6*z + 9 es un polinomio, por lo que está definido para todo z real.
print("\nb) Puntos donde h'(z) no está definida:")
print("h'(z) es un polinomio y está definido para todos los z reales. No hay puntos críticos de este tipo.")
singularities_h_prime = [] # Lista vacía

# Combinamos todos los puntos críticos (reales)
all_critical_points_values_h = sorted(list(set(critical_points_h_prime_zero + singularities_h_prime)))
print("\nTodos los puntos críticos (valores reales de z):")
if not all_critical_points_values_h:
    print("No hay puntos críticos reales.")
else:
    try:
        display(Math(f"z_{{crit}} = {sympy.latex(all_critical_points_values_h)}"))
    except NameError:
        sympy.pprint(all_critical_points_values_h)
```

```python
# TERCER EJERCICIO

# --- PASO 1: Definir el símbolo y la función ---
print("--- PASO 1: Definir el símbolo y la función ---")
x = sympy.symbols('x')
f_expr = sympy.sqrt(4 - x**2)

# Mostramos la función (usando display y Math para formato LaTeX si está disponible)
display(Math(f"f(x) = {sympy.latex(f_expr)}"))

# --- PASO 2: Determinar el dominio de la función ---
# La expresión dentro de la raíz cuadrada debe ser no negativa: 4 - x^2 >= 0
print("\n--- PASO 2: Determinar el dominio de la función ---")
domain_condition = 4 - x**2 >= 0
# Usamos sympy.S.Reals para especificar que buscamos soluciones en los números reales
domain_f = sympy.solveset(domain_condition, x, domain=sympy.S.Reals) # S.Reals es un singleton de SymPy
display(Math(f"\\text{{Dominio de }} f(x): D = {sympy.latex(domain_f)}"))

# --- PASO 3: Calcular la primera derivada (f'(x)) ---
print("\n--- PASO 3: Calcular la primera derivada f'(x) ---")
f_prime = sympy.diff(f_expr, x)
display(Math(f"f'(x) = {sympy.latex(f_prime)}"))

# --- PASO 4: Encontrar los puntos críticos ---
# Los puntos críticos ocurren donde f'(x) = 0 o donde f'(x) no existe.
print("\n--- PASO 4: Encontrar los puntos críticos ---")

# a) Puntos donde f'(x) = 0
print("\na) Puntos donde f'(x) = 0:")
# Usamos sympy.solve para encontrar las raíces de f_prime
critical_points_df_zero_raw = sympy.solve(f_prime, x)
# Filtramos para asegurarnos de que están en el dominio
critical_points_df_zero = [p for p in critical_points_df_zero_raw if domain_f.contains(p)]
display(Math(f"\\text{{Soluciones para }} f'(x) = 0: x = {sympy.latex(critical_points_df_zero)}"))


# b) Puntos donde f'(x) no está definida
# La derivada f'(x) = -x / sqrt(4 - x^2) no está definida cuando el denominador es 0.
print("\nb) Puntos donde f'(x) no está definida:")
# El denominador de f_prime es sympy.sqrt(4 - x**2)
# No está definido cuando 4 - x**2 = 0
denominator_expr_f_prime = sympy.denom(f_prime) # Sympy puede intentar extraerlo
# Más directamente, sabemos que es sqrt(4-x**2), por lo que el problema es cuando 4-x**2 = 0
denominator_zero_condition = 4 - x**2
singularities_df_raw = sympy.solve(denominator_zero_condition, x)
# Filtramos para asegurarnos de que están en el dominio
singularities_df = [p for p in singularities_df_raw if domain_f.contains(p)]
display(
    Math(f"\\text{{Puntos donde }} f'(x) \\text{{ no existe (denominador de }} f'(x) = 0 \\text{{ y están en el dominio): }} x = {sympy.latex(singularities_df)}")
    )

# Combinamos todos los puntos críticos
all_critical_points_values = sorted(list(set(critical_points_df_zero + singularities_df)))
print("\nTodos los puntos críticos (valores de x):")
display(Math(f"x_{{crit}} = {sympy.latex(all_critical_points_values)}"))

```

```python
# PRIMER EJERCICIO 

# Definir el símbolo
x = sympy.symbols('x')

# Definir el integrando por separado
integrand1 = (8*x - 12)*(4*x**2 - 12*x)**4
integrand2 = 1/x

# Calcular las integrales por separado
integral_part1_expanded = sympy.integrate(integrand1, x)
integral_part2 = sympy.integrate(integrand2, x)

# Factorizar el resultado de la primera parte para obtener la forma compacta
integral_part1_compact = sympy.factor(integral_part1_expanded)
print(f"Integral de la primera parte (factorizada): {integral_part1_compact}")

# Imprimir las partes
print(f"Integral de la primera parte: {integral_part1_compact}")
print(f"Integral de la segunda parte: {integral_part2}")

# Combinar los resultados
resultado_compacto = integral_part1_compact + integral_part2

# Imprimir el resultado
print(f"\nResultado combinado (forma compacta): {resultado_compacto} + C")
display(resultado_compacto)

# Simplificar el resultado
integral_result_simplified = sympy.simplify(resultado_compacto)
print(f"\nForma simplificada de la integral:")
display(integral_result_simplified)
```

```python
# SEGUNDO EJERCICIO 

# Definir los símbolos
x = sympy.symbols('x')
e = sympy.E # Constante de Euler

# Definir la función a integrar
f = x**2 * e**x
print('Función a integrar f(x) =', f)
display(f)

# Calcular la integral indefinida
integral_f = sympy.integrate(f, x)

# Imprimir el resultado
print(f"La integral de {f} es: {integral_f} + C")

# Simplificar la expresión 
resultado_manual_simplificado = sympy.factor(x**2 * e**x - 2*x*e**x + 2*e**x)
print(f"Resultado manual simplificado: {resultado_manual_simplificado} + C")
display(resultado_manual_simplificado + sympy.symbols('C'))
```
