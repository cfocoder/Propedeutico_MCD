## The Plot Function

### Required Parameters
* Expression: The SymPy expression to plot
* Range: Tuple (x, xmin, xmax) specifying:
    * x: Symbol variable for the expression
    * xmin, xmax: Start and end values for the x-axis

### Optional Parameters
#### Display Control
* show (bool, default=True): Whether to display the plot immediately
* title (str): Title of the plot
* xlabel, ylabel (str): Axis labels
* xlim, ylim (tuple): Axis limits as (min, max)

#### Style Parameters
* line_color (str or RGB tuple): Color of the line (e.g., 'red', '#FF0000')
* line_width (float, default=1.0): Width of the plotted line
* adaptive (bool, default=True): Whether to use adaptive sampling
* depth (int): For adaptive sampling, level of recursion
* nb_of_points (int, default=100): Number of points when not using adaptive sampling

#### Legend and Grid
* legend (bool, default=False): Whether to show a legend
* grid (bool, default=False): Whether to show a grid
* annotations (list): List of annotation objects to add

#### Formatting
* xscale, yscale (str): Scale type ('linear', 'log', etc.)
* axis_center (tuple or bool): Position of axis crossing
* aspect_ratio (str or tuple): Control aspect ratio
* size (tuple): Size of the figure as (width, height) in pixels

#### Backend Options
* backend (str, default='matplotlib'): Plotting backend to use
* backend_options (dict): Options to pass to the backend

```python
import sympy as sp
from sympy import symbols, sin, cos, pi
from sympy.plotting import plot

# Define symbolic variable
x = symbols('x')

# Create expressions to plot
expr1 = sin(x)
expr2 = sin(x + pi/4)
expr3 = x**2

# Create a 2D plot with multiple functions
# Instead of passing line_color as a list, we'll create individual plot objects
p1 = plot(expr1, (x, -2*pi, 2*pi), show=False, label='sin(x)', line_color='blue')
p2 = plot(expr2, (x, -2*pi, 2*pi), show=False, label='sin(x + /4)', line_color='red')
p3 = plot(expr3, (x, -2, 2), show=False, label='x^2', line_color='green')

# Extend the plots to combine them
p1.extend(p2)
p1.extend(p3)

# Add title and labels
p1.title = 'SymPy 2D Plot Example'
p1.xlabel = 'x axis'
p1.ylabel = 'y axis'
p1.legend = True

# Display the plot
p1.show()

# To save the plot to a file:
# p1.save('sympy_2d_plot.png')
```

```python
import sympy as sp
from sympy import symbols, solve, plot
import numpy as np

# Define symbolic variables
x, y = symbols('x y')

# Define two linear equations
line1 = 2*x + 1  # y = 2x + 1
line2 = -x + 4   # y = -x + 4

# Find the intersection point by solving the equation
# 2x + 1 = -x + 4
intersection_x = solve(line1 - line2, x)[0]
intersection_y = line1.subs(x, intersection_x)

print(f"Intersection point: ({intersection_x}, {intersection_y})")

# Create a plot of the two lines
p1 = plot(line1, (x, -3, 3), show=False, label='y = 2x + 1', line_color='blue')
p2 = plot(line2, (x, -3, 3), show=False, label='y = -x + 4', line_color='red')

# Combine the plots
p1.extend(p2)

# Add title and labels
p1.title = 'Two Intersecting Lines'
p1.xlabel = 'x'
p1.ylabel = 'y'
p1.legend = True

# Mark the intersection point
# Note: SymPy's plot doesn't easily support scatter points
# We can add annotations or use matplotlib directly for more control

# Show the plot
p1.show()
```

```python
import sympy as sp
from sympy import symbols, sin, cos, exp, pi
from sympy.plotting import plot3d, plot3d_parametric_surface

# Define symbolic variables
x, y = symbols('x y')

# Example 1: Simple 3D surface plot
# Plot z = sin(x) * cos(y)
expr1 = sin(x) * cos(y)
p1 = plot3d(
    expr1, (x, -3, 3), (y, -3, 3),
    title='3D Surface: sin(x) * cos(y)',
    xlabel='x', ylabel='y', zlabel='z'
)

# Example 2: More complex expression
# Plot z = sin(x*y) / (x*y)
expr2 = sin(x*y) / (x*y + 0.1)  # Adding 0.1 to avoid division by zero
p2 = plot3d(
    expr2, (x, -5, 5), (y, -5, 5),
    title='3D Surface: sin(x*y)/(x*y+0.1)'
)

# Example 3: Parametric surface (a torus)
# Parametric equations for a torus
u, v = symbols('u v')
r, R = 1, 3  # Minor and major radii
x_expr = (R + r*cos(v))*cos(u)
y_expr = (R + r*cos(v))*sin(u)
z_expr = r*sin(v)

p3 = plot3d_parametric_surface(
    x_expr, y_expr, z_expr,
    (u, 0, 2*pi), (v, 0, 2*pi),
    title='Parametric Surface: Torus'
)
```
