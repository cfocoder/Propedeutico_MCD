## Vectores

En el contexto de la ciencia de datos, un **vector** es una estructura matemática fundamental que representa datos, observaciones o características en forma de lista ordenada de números. Matemáticamente se trata de un arreglo unidimensional de números reales (o complejos), y se denota típicamente como:

\begin{equation}
\mathbf{x} = 
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
\quad \text{o bien} \quad
\mathbf{x} = [x_1, x_2, \dots, x_n]
\end{equation}

Donde cada componente $x_i$ representa una característica o valor numérico.

Un vector puede pensarse como un **punto en el espacio**, una **dirección y magnitud**, o simplemente como una **colección de características** de un objeto.

#### Ejemplo

En un dataset de pacientes, podrías tener un vector que represente a una persona como:

$$
\mathbf{x} = [25, 1.70, 68]
$$

donde:
- 25 es la edad (años),
- 1.70 es la estatura (m),
- 68 es el peso (kg).

### Interpretación geométrica

- En 2D o 3D, un vector puede visualizarse como una flecha desde el origen hasta el punto que define.
- La **magnitud** (norma) del vector indica qué tan grande es el punto en el espacio.
- Dos vectores pueden compararse en términos de **dirección y ángulo**.

### Importancia en Ciencia de Datos

- **Observaciones individuales** en un dataset suelen representarse como vectores.
- **Filas de una matriz de datos** son vectores de características.
- Se utilizan en:
  - Cálculo de **distancias** entre puntos (por ejemplo, en clustering o k-NN).
  - **Modelos lineales** como regresión.
  - **Transformaciones** en PCA y reducción de dimensionalidad.
  - Representaciones en **espacios de características** (por ejemplo, embeddings en NLP).
  - **Álgebra lineal computacional** con librerías como NumPy.

## OLS utilizando matrices

### Para una sola variable dependiente

El modelo lineal se expresa de forma matricial por $Y = X \beta + \epsilon$ (cada variables es una matriz), expresada de la siguiente forma:

\begin{equation}
\begin{pmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{pmatrix} =
\begin{pmatrix}
1 & x_{11} & x_{12} & \cdots & x_{1q} \\
1 & x_{21} & x_{22} & \cdots & x_{2q} \\
\vdots & \vdots & \vdots & & \vdots \\
1 & x_{n1} & x_{n2} & \cdots & x_{nq} \\
\end{pmatrix}
\begin{pmatrix}
\beta_0 \\
\beta_1 \\
\vdots \\
\beta_q
\end{pmatrix}
+
\begin{pmatrix}
\varepsilon_1 \\
\varepsilon_2 \\
\vdots \\
\varepsilon_n
\end{pmatrix}
\end{equation}

El número de filas $n$ corresponde al número de observaciones de los datos (en este caso 85 filas). El número de variables de entrada está dado por $q$, y este valor nos resulta en el número de variables a ajustar para el modelo lineal $\beta_0, \beta_1 \ldots \beta_q$. Por lo tanto, las dimensiones de las matrices son las siguientes, $Y (n \times 1)$, $X (n \times (q+1))$, $\beta ((q+1) \times 1)$, $\epsilon (n \times 1)$. Recordar que una operación entre matrices tiene que tomar en cuenta las dimensiones entre ellas, y normalmente no es conmutativa $(XY \ne YX)$.

**Solución al problema de regresión lineal múltiple por mínimos cuadrados ordinarios (OLS)**
En un problema de regresión lineal multivariable, el vector de parámetros óptimos $\beta$ se encuentra utilizando normalmente el método de mínimos cuadrados ordinarios (OLS), utilizando la siguiente ecuación,

\begin{equation}
\beta = (X^TX)^{-1}X^TY
\end{equation}

Podemos ir construyendo estas matrices utilizando la librería de `numpy`en python.

```python
# Creacion de la matriz X

# Cremos una lista de 1's que tenga la misma longitud que los datos
x1 = []
for i in range(0,data.shape[0]):
    x1.append(1.0)

# Creamos lista con la primer variable
x2 = data['TV'].values.tolist()

# Creamos la matriz X
X = []
for i in range(0,data.shape[0]):
    X.append([x1[i],x2[i]])
    print([x1[i],x2[i]])
    

#print(X)
```

```python
# Creamos Matriz Y
y = data['Sales'].values.tolist()
Y = []
for i in range(0,data.shape[0]):
        Y.append([y[i]])
        print([y[i]])


#print(Y)
```

A continuación, calculamos la matriz transpuesta $\boldsymbol{X}^T$. Recordar que una matriz transpuesta se intercambia el número de columnas y de filas.

```python
# Calculamos la matriz t con numpy
X_T = np.transpose(X)

print(X_T)
```

Necesitamos calcular $X^T X$, para esto usamos el módulo `numpy.dot`. Recordar que aquí el orden de las variables (matrices) es importante

```python
M = X_T.dot(X) # Este producto de matrices se guarda en una variable M

print(M)
```

Necesitamos calcular $(X^T X)^{-1}$. Necesitamos sacar una matriz inversa

```python
# calculamos matriz inversa con numpy
M_i = np.linalg.inv(M)

print(M_i)
```

```python
# Multiplicar una matriz por su inversa nos tiene que dar una matriz identidad (diagonal de 1's)

print(M.dot(M_i))

# Si nos da una matriz identidad
```

Ahora tenemos que ir multiplicando producto por producto, Hasta ahora hemos guardado $(X^T X)^{-1}$ en una variable M_i, de tal forma que $\beta = M_i X^T Y$, tenemos que ir haciendo producto por producto (en orden)

```python
# primero X_T * Y

B = X_T.dot(Y)
print(B)
```

```python
#Luego M_i * (X_T * Y)

Beta = M_i.dot(B)
```

```python
print(Beta)
```
